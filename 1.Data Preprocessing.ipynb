# Changing the data type

```python
#Transforming categorical numbers into string
dataset["ID"]=dataset["ID"].astype('str')
dataset["Inception"]=dataset["Inception"].astype('str')
```

```python
dataset.info()
```

```python
#Transform percentage(string) into a fraction
dataset['Growth'] = dataset['Growth'].str.rstrip('%').astype('float') / 100.0
```

```python
#Transform revenue into a float
dataset['Revenue'] = dataset['Revenue'].str.replace('$','')
dataset['Revenue'] = dataset['Revenue'].str.replace(',','').astype('float')
```

```python
#Transform expenses into a float
dataset['Expenses'] = dataset['Expenses'].str.replace(' Dollars','')
dataset['Expenses'] = dataset['Expenses'].str.replace(',','').astype('float')
```

```python
dataset.info()
```

```python
#Checking for obvious anormalities
dataset.describe()
```

```python
#looking for data that wouldn't make any sense
dataset[dataset['Expenses'] <= 0]
```
